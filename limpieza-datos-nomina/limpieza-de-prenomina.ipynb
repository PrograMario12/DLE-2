```python
scaler = StandardScaler()
prenomina_normalized = prenomina_horizontal.copy()

# Seleccionar las columnas numéricas para normalizar
numeric_columns = prenomina_normalized.select_dtypes(include=[np.number]).columns

# Normalizar las columnas numéricas
prenomina_normalized[numeric_columns] = scaler.fit_transform(prenomina_normalized[numeric_columns])

# Verificar el resultado
display(prenomina_normalized.head())
```

```python
columns_to_keep = ['# Empleado', 'Nombre', 'Departamento', 'Linea', 'Centro de Costos', 'T. Hrs', 'Asistencia']
prenomina_horizontal = prenomina_horizontal[columns_to_keep]
```

```python
display(prenomina_horizontal.head())
```

```python
import os

file_name = 'prenomina_horizontal_gold.csv'

# Check if the file already exists
if os.path.exists(file_name):
    # Delete the existing file
    os.remove(file_name)

# Create the file
prenomina_horizontal.to_csv(file_name, index=False)
```
